```python
# O sintetizador de voz pode ser visto como uma função em python
# que recebe a frase, o dicionário de sílabas e o dicionário de fonemas
# e devolve um vetor numpy com o áudio final da frase

def sintetizador_voz(frase, dicionario_silabas, dicionario_fonemas):
    audio_sintetizado = np.array([])
    palavras = frase.split()
    for palavra in palavras:
        silabas = dicionario_silabas.get(palavra)
        if silabas:
            for silaba in silabas:
                audio_silaba, _ = librosa.load(dicionario_fonemas.get(silaba))
                audio_sintetizado = np.concatenate((audio_sintetizado, audio_silaba))
    return audio_sintetizado
```

### d) Faça quatro gráficos boxplot. <font color='red'>(1 ponto)</font>
Da **idade (age)**, do **valor (value)**, do **salário (wage)** e do **overall** dos jogadores.

Pinte cada gráfico boxplot com a coluna **Performance** realizada no item c (parâmetro hue).

O que você pode interepretar desses gráficos? 

```python
# Coloque seu código aqui
#axes[1].set_yscale('log') Pode ajudar a modificar a escala do eixo y quando os valores são muito grandes

caminho = r'data_fifa2019.csv'
data_fifa = pd.read_csv(caminho)

colunas_selecionadas = ['Name', 'Age', 'Nationality', 'Club', 'Overall', 'Value', 'Wage', 'Position']
data_fifa = data_fifa[colunas_selecionadas]

def converter_valores(value):
    if value[-1] == 'M':
        return float(value[1:-1]) * 1000000
    elif value[-1] == 'K':
        return float(value[1:-1]) * 1000
    else:
        return float(value[1:])

data_fifa['Value'] = data_fifa['Value'].apply(converter_valores)
data_fifa['Wage'] = data_fifa['Wage'].apply(converter_valores)

def overall_classes(x):
    if x < 60:
        return 'Baixa'
    if x < 75:
        return 'Normal'
    return 'Alta'

data_fifa['Performance'] = data_fifa['Overall'].apply(overall_classes)

graficos = ["Age", "Value", "Wage", "Overall"]

fig, axs = plt.subplots(1, len(graficos), figsize=(20, 5))

for i, grafico in enumerate(graficos):
    sns.boxplot(x="Performance", y=grafico, data=data_fifa, ax=axs[i])
    if grafico in ["Wage", "Value"]:
        axs[i].set_yscale("log")
    axs[i].set_title(grafico + " por Performance", fontsize=14)
    axs[i].set_xlabel("Performance", fontsize=12)
    axs[i].set_ylabel(grafico, fontsize=12)
    axs[i].tick_params(labelsize=10)

plt.show()

# Os gráficos indicam que a performance dos jogadores está diretamente relacionada ao seu salário e valor, mas não há uma forte relação com sua idade.
# Jogadores com performance "Alta" têm idade semelhante aos de performance "Normal", mas são mais velhos do que os de performance "Baixa".
# A performance está correlacionada com o valor dos jogadores, com os jogadores de performance "Alta" apresentando um valor médio mais elevado em comparação aos demais.
# Jogadores de performance "Alta" têm salários médios significativamente maiores em comparação aos jogadores com desempenho inferior.
```
